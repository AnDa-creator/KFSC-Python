```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### d variation

```python
file_name = "PokerHand2.matd_vary_run_info.csv"
Score_df = pd.read_csv(file_name, header=None)[0:24][:]
Score_df
```

```python
acc_linestyle = ['b--', 'k:', 'g-', 'r-.', 'c--']
```

```python
dataset_name = Score_df[0][0]
dataset_name = dataset_name[0:9]
acc_values = Score_df[:][1].to_numpy()
d_values = Score_df[:][4].to_numpy()
nmi_values = Score_df[:][2].to_numpy()
time_values = Score_df[:][3].to_numpy()
function_name = Score_df[0][7]
```

```python
plt.plot(d_values, acc_values, acc_linestyle[0], label=f"Accuracy(%) ," + dataset_name)
plt.plot(d_values, nmi_values, acc_linestyle[1], label=f"Normal mutual Info(%) ," + dataset_name)
# plt.plot(det_ax_2*1e3, np.abs(curr_det_axis_2)*1e-3, I_linestyle[1], label="$V_{sd} = -5 mV$ , $t_{so} = 0$")

plt.ylabel(f'Value (%)',fontsize=16)
plt.xlabel(f'Parameter d , dictionary feature size',fontsize=16)
plt.legend(loc='best',ncol=2, fancybox=True, shadow=True, bbox_to_anchor=(-0.05, 0.4, 1, 0.75))
plt.tick_params(which='both', direction='in', labelsize=14)
plt.grid(linestyle = '--', linewidth = 0.5)
plt.tight_layout()
plt.savefig( dataset_name+ 'Dvary' + '.png', bbox_inches='tight', dpi = 600)
plt.show()
```

### Batch size Variation in KFSC-MB (p)

```python
file_name = "mnistbatch_vary_run_info.csv"
Score_df = pd.read_csv(file_name, header=None)[0:46][:]
Score_df
```

```python
acc_linestyle = ['b--', 'k:', 'g-', 'r-.', 'c--']
```

```python
eval(Score_df[:][8][0])
```

```python
dataset_name = Score_df[0][0]
dataset_name = dataset_name[0:6]
acc_values = Score_df[:][1].to_numpy()
batch_values = np.array([eval(Score_df[:][8][i])['opt']['batch_size'] for i in range(len(acc_values))])
nmi_values = Score_df[:][2].to_numpy()
time_values = Score_df[:][3].to_numpy()
function_name = Score_df[0][7]
batch_values
```

```python
plt.plot(batch_values, acc_values, acc_linestyle[4], label=f"Accuracy(%) ," + dataset_name)
plt.plot(batch_values, nmi_values, acc_linestyle[5], label=f"Normal mutual Info(%) ," + dataset_name)
# plt.plot(det_ax_2*1e3, np.abs(curr_det_axis_2)*1e-3, I_linestyle[1], label="$V_{sd} = -5 mV$ , $t_{so} = 0$")

plt.ylabel(f'Value (%)',fontsize=16)
plt.xlabel(f'Parameter nsel , Batch size for KFSC-L',fontsize=16)
plt.legend(loc='best',ncol=2, fancybox=True, shadow=True, bbox_to_anchor=(-0.05, 0.4, 1, 0.75))
plt.tick_params(which='both', direction='in', labelsize=14)
plt.grid(linestyle = '--', linewidth = 0.5)
plt.tight_layout()
plt.xscale('log')
plt.savefig( dataset_name+ 'batchvary' + '.png', bbox_inches='tight', dpi = 600)
plt.show()
```
